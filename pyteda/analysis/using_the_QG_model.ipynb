```python
import numpy as np
import matplotlib.pyplot as plt
from pyteda.models import QGModel

# --- Parmetros ---
N = 18
dt = 0.01
tend = 5
F = 1600.0
nt = int(tend / dt)

# --- Inicializacin del modelo ---
model = QGModel(N=N, dt=dt, F=F)
T = np.linspace(0, tend, nt + 1)

# --- Condicin inicial (se genera y se propaga internamente) ---
x0 = model.get_initial_condition(seed=42,)  #  y q
snapshots = []

# --- Repropagamos manualmente (sin anlisis ni observaciones) ---
x = x0.copy()
for step in range(nt):
    x = model.propagate(x, T[:2])  # solo un paso corto
    T[:2] += dt  # avanzar T

    if step in [0, nt // 2, nt - 1]:
        q = x[:N*N].reshape(N, N)
        psi = x[N*N:].reshape(N, N)
        snapshots.append((psi.copy(), model.laplacian(psi).copy()))

# --- Visualizacin ---
x = np.linspace(0, 1, N, endpoint=False)
y = np.linspace(0, 1, N, endpoint=False)
X, Y = np.meshgrid(x, y)

fig, axes = plt.subplots(2, 3, figsize=(18, 7))
labels = ['Inicial', 'Mitad', 'Final']

for i, (psi_snap, zeta_snap) in enumerate(snapshots):
    im1 = axes[0, i].contourf(X, Y, psi_snap, levels=40, cmap='RdBu_r')
    axes[0, i].set_title(f'Funcin de corriente () - {labels[i]}')
    plt.colorbar(im1, ax=axes[0, i])

    im2 = axes[1, i].contourf(X, Y, zeta_snap, levels=40, cmap='PuOr')
    axes[1, i].set_title(f'Vorticidad relativa () - {labels[i]}')
    plt.colorbar(im2, ax=axes[1, i])

for ax in axes.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```
